```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uber_df = pd.read_csv('uber.csv')
uber_df.head()
```

```python
uber_df.rename(columns = {'Date/Time' : 'DateTime'}, inplace = True)
```

```python
uber_df['DateTime'] = pd.to_datetime(uber_df['DateTime'])
```

```python
uber_df['Day'] = uber_df['DateTime'].apply(lambda x: x.day)
```

```python
uber_df['Weekday'] = uber_df['DateTime'].apply(lambda x: x.weekday())
```

```python
uber_df['Hour'] = uber_df['DateTime'].apply(lambda x: x.hour)
```

```python
plt.figure(figsize = (16,9))
sns.distplot(uber_df['Day'])
```

```python
plt.figure(figsize = (16,9))
sns.distplot(uber_df['Hour'])
```

```python
plt.figure(figsize = (16,9))
sns.distplot(uber_df['Weekday'])
```

```python
corr_df = uber_df.groupby(['Weekday', 'Hour']).apply(lambda x: len(x))
corr_df = corr_df.unstack()
plt.figure(figsize = (16,9))
sns.heatmap(corr_df)
```

```python
plt.figure(figsize = (16,9))
plt.title('Uber Trip Analysis', size = 12)
sns.scatterplot(uber_df, x='Lon', y='Lat')
```
