#### 4. Correlation heatmap between variable

```python
# Select only numeric columns (assuming all non-numeric columns should be excluded)
numeric_columns = df_tracks.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_df = numeric_columns.corr(method='pearson')
```

```python
plt.figure(figsize=(12, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title("Correlation Heatmap Between Variables")
```

#### Remove rows with invalid date values

```python
df_tracks = df_tracks[df_tracks['release_date'].str.match(r'\d{4}-\d{2}-\d{2}')]
```

#### Set the release date column as the index

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
```

#### Create a new column called years from release date column

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

#### 7. Distribution plot (bar plot) for total number of song per year

```python
plt.figure(figsize=(10,5))
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of songs per year')
```

#### Convert milliseconds to seconds

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
```

#### 8. Duration of the songs over the years

```python
total_dr = df_tracks.duration
```

```python
fig, ax = plt.subplots(figsize=(18, 8))
sns.barplot(x=years, y=total_dr, ax=ax)
ax.set(title='Year vs Duration')
plt.xticks(rotation=90)
ax.set_xlabel('Year')  # Add x-axis label
```

#### 9. Line plot for average duration of the songs over the years

```python
total_dr = df_tracks.duration
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=years, y=total_dr)
plt.title('Year vs Duration')
plt.xlabel('Year')
```

#### Load the second dataset

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

#### 10. Duration of the song for different genres

```python
plt.figure(figsize=(10, 5))
sns.color_palette('rocket', as_cmap = True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.title('Duration of the songs in different genres')
plt.xlabel("Duration in milli seconds")
plt.ylabel('Genres')
```

#### 11. Top 5 genres by popularity

```python
famous = df_genre.sort_values('popularity', ascending = False).head(10)
plt.figure(figsize = (10,5))
sns.set_style(style = 'darkgrid')
sns.barplot(y='genre', x='popularity', data= famous).set(title = "Top 5 genres by popularity")
```
